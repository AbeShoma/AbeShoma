# Exercise 1: Write a Python program to read the sensor locations and show them on a map

### The Pandas library is imported as pd and the Plotly Express library as px. These libraries are used for data processing and visualization.

```python
import pandas as pd
import plotly.express as px
```

### Use the pd.read_csv() function to read data from a CSV file and store it in a Pandas DataFrame called df_stationinfo.



```python
df_stationinfo = pd.read_csv("./station_Info.csv")
```

### Data is taken from the 'Location' column and stored in a list called Location. Next, it creates a list containing the latitude and longitude values by splitting each element with a space and parentheses. This list is reassigned to the Location variable.

```python
#Store data from location to list
Location = list(df_stationinfo['Location'])
Location = [l.split('(')[1].split(')')[0].split(' ')for l in Location]
```

### Creates a list of latitude and longitude from the Location list, converting it to a float type.

```python
#Get a list of latitude and longitude
latitude = [float(l[1]) for l in Location]
longitude = [float(l[0]) for l in Location]
```

### Use the latitude and longitude lists and add them as new columns to the df_stationinfo DataFrame. This will include the latitude and longitude information in the original dataframe.

```python
#Add latitude and longitude to data frame
df_stationinfo['latitude'] = latitude
df_stationinfo['longitude'] = longitude
```

### Plotting a scatter plot on a map using plotly.express and mapbox. df_stationinfo data frame.' lat' is latitude, 'lon' is longitude, and hove_name is information about the point to be displayed when hovering, coordinates, etc.

```python
fig = px.scatter_mapbox(
    data_frame = df_stationinfo,
    lat = "latitude",
    lon = "longitude",

    hover_name = "Address",
    hover_data = ["Location"],

    center={'lat':34.686567, 'lon':135.52000},
    zoom=4,
    height=600,
    width=800
)
```

#### Set and display map style, margins, and title.

```python
fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.update_layout(title_text="Station place")
fig.show()
```
